```python
#import the library 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset

df = pd.read_csv("student_feedback.csv")
df.head(10)
```

```python
#data Understanding & cleaning

df.info()
df.isnull().sum()
```

```python
#define feedback columns 

feedback_cols= [
    "Well versed with the subject",
    "Explains concepts in an understandable way",
    "Use of presentations",
    "Degree of difficulty of assignments",
    "Solves doubts willingly",
    "Structuring of the course",
    "Provides support for students going above and beyond",
    "Course recommendation based on relevance"
]
```

```python
#AVERAGE RATING PER PARAMETER 

avg_ratings = df[feedback_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index)
plt.title("Average Student Ratings by Parameter")
plt.xlabel("Average Rating")
plt.ylabel("Feedback Parameter")
plt.show()
```

```python
#Overall Course Satisfaction Score (KPI)

overall_score = df[feedback_cols].mean().mean()
overall_score
```

```python
#Rating Distribution (Student Opinion Consistency)

plt.figure(figsize=(12,6))
sns.boxplot(data=df[feedback_cols])
plt.xticks(rotation=45, ha="right")
plt.title("Distribution of Student Ratings")
plt.show()
```

```python
#Teaching Quality VS Assignment Difficulty

compare_df = df[
    ["Explains concepts in an understandable way",
     "Degree of difficulty of assignments"]
].mean()

compare_df.plot(kind="bar", figsize=(6,4))
plt.title("Teaching Clarity vs Assignment Difficulty")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Course Recommendation Analysis 

plt.figure(figsize=(5,4))
sns.histplot(df["Course recommendation based on relevance"], bins=5, kde=True)
plt.title("Course Recommendation Distribution")
plt.xlabel("Ratings")
plt.show()
```
